```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"Coffee Shop Sales.xlsx")
```

```python
df.drop(['transaction_id','product_id'], axis = 1,inplace = True)
```

```python
df['trans_year'] = df['transaction_date'].dt.year
df.sample(10)
```

```python
df['trans_month'] = df['transaction_date'].dt.month_name()
df.sample(10)
```

```python
df['weekend'] = np.where(df['transaction_date'].dt.dayofweek > 4, 1, 0)
df.sample(10)
```

```python
sns.countplot(x='weekend', data=df)
```

```python
sns.countplot(x='trans_month', data=df)
```

```python
df['transaction_datetime'] = pd.to_datetime(df['transaction_date'].astype(str) + ' ' + df['transaction_time'].astype(str))
```

```python
df['hour'] = pd.to_datetime(df['transaction_datetime']).dt.hour
```

```python
sns.countplot(x='hour', data=df)
```

**Peak Hours and Staffing**:

**Problem**: High transaction frequency during early morning hours could overwhelm staff.

**Solution** : Increase staffing during peak times to improve service efficiency.


```python
plt.pie(df['product_category'].value_counts(), labels=df['product_category'].unique())
plt.figure(figsize=(30,10))
plt.show()
```

**Revenue Contribution by Category:**

**Problem**: Significant revenue from categories like "Coffee" and "Tea"; less from others.

**Solution**: Focus marketing on high-revenue categories and evaluate low-revenue categories for potential improvements or removal.


```python
sns.countplot(x='product_category', data=df, order=df['product_category'].value_counts().iloc[:5].index)  # Limit to top 5 categories
plt.figure(figsize=(25,10))
plt.show()
```

**Product Popularity and Inventory Management:**

**Problem**: Popular items (e.g., Coffee,Tea ) may frequently sell out.

**Solution**: Ensure adequate stock levels of high-demand products.
